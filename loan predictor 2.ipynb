```python
y=df.iloc[:,[-1]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder,MinMaxScaler
from sklearn.impute import SimpleImputer
from sklearn.model_selection import GridSearchCV
```

```python
# oh1=OneHotEncoder(sparse_output=False,drop="first",handle_unknown="ignore")
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
# trf1=ColumnTransformer(transformers=[
#     ("si1",SimpleImputer(strategy="mean"),[5,6,7])
# ]
#                        ,remainder="passthrough"  
# )
```

```python
# trf2 = ColumnTransformer(transformers=[
#     ("ohe",oh1,[1,2])
#      ]
#                        ,remainder='passthrough')
```

```python
trf3=ColumnTransformer(transformers=[
    ("min",MinMaxScaler(),slice(0,10))
]
    ,remainder="passthrough"
)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
trf4=RandomForestClassifier(n_estimators=10,bootstrap=True,n_jobs=-1,oob_score=True,max_depth=5)
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline([
    
    # ("trf2",trf2),
    ("trf3",trf3),
    ("trf4",trf4)
   
]
)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
